```python
%load_ext autoreload
import numpy as np
import astropy.units as units
```

```python
%autoreload 2
import dualpy as dp
```

```python
a = (np.arange(10)) * units.s
b = (np.arange(100) + 100) * units.s
c = (np.arange(1) + 1000) * units.s
```

```python
def remake(x):
    return np.broadcast_to(x[:,np.newaxis], shape=[x.size,5]) * x.unit
a = remake(a)
b = remake(b)
c = remake(c)
```

```python
aa = dp.seed(a, "a")
bb = dp.seed(b, "b")
cc = dp.seed(c, "c")
```

```python
xx = np.concatenate((aa,bb,cc))
```

```python
print(xx)
```

```python
print(xx.jacobians['b'].todensearray())
```
